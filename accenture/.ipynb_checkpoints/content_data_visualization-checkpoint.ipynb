# Data Analysis and Visualization

This Python script performs data analysis and visualization on a dataset containing information about content reactions. The dataset includes details such as content ID, reaction type, datetime, content type, category, sentiment, and score.

## Purpose
The purpose of this analysis is to gain insights into the performance of different content categories based on user reactions. By analyzing the data, we aim to:

- Identify the top 5 performing content categories.
- Visualize the distribution of reactions across these categories.
- Provide additional insights such as the number of unique categories, reactions to the most popular category, and the month with the most posts.

## Methodology
1. **Data Cleaning**: The dataset is cleaned by removing duplicates and handling missing values.
2. **Data Merging**: The cleaned datasets are merged to create a final dataset with all relevant information.
3. **Visualization**: Matplotlib library is used to create visualizations, including pie charts and bar charts, to represent the top performing categories and other insights.
4. **Analysis**: Various statistical measures are calculated to provide additional insights into the dataset.

## Results
The analysis provides valuable insights into the performance of content categories and user reactions. These insights can be used to optimize content strategies and improve user engagement.



```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
final_data = pd.read_csv('top_5_categories.csv')

# Pie chart for top 5 content categories
top_categories = final_data['Category'].value_counts().head(5)
plt.figure(figsize=(8, 8))
plt.pie(top_categories, labels=top_categories.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Content Categories')
plt.axis('equal')
plt.show()

# Bar chart for the total score of top 5 content categories
category_scores = final_data.groupby('Category')['Score'].sum().sort_values(ascending=False).head(5)
plt.figure(figsize=(10, 6))
category_scores.plot(kind='bar', color='skyblue')
plt.title('Top 5 Content Categories by Total Score')
plt.xlabel('Category')
plt.ylabel('Total Score')
plt.xticks(rotation=45, ha='right')
plt.show()

# Unique categories count
unique_categories_count = final_data['Category'].nunique()
print("Number of Unique Categories:", unique_categories_count)

# Number of reactions to the most popular category
most_popular_category = final_data['Category'].value_counts().idxmax()
reactions_to_most_popular = final_data[final_data['Category'] == most_popular_category].shape[0]
print("Number of Reactions to the Most Popular Category:", reactions_to_most_popular)

# Month with the most posts
final_data['Datetime'] = pd.to_datetime(final_data['Datetime'])
final_data['Month'] = final_data['Datetime'].dt.month
month_with_most_posts = final_data['Month'].value_counts().idxmax()
print("Month with the Most Posts:", month_with_most_posts)
```
